```python
# increasing l increases the correlation between nearby points and yields a smoother curve.
# decreasing l means that nearby points are less correlated and the curves get less smooth.
# l = 100000 effectively is a flat, horizontal line.
# sigma varies the scale of the kernel, and thus the covariance and effectively the range of the gaussian.
# interestingly, decreasing sigma reduces the smoothness of the curve, along with the range. this is while we keep l constant. hmmm.
# changing l doesn't make much of a difference. but in some non-quantifiable way, decreasing l fucks it up even more than increasing it.
# the eye says so, at least.
```

```python
def se_kernel(a, b, l = 1, sigma = 1):
    """ GP squared exponential kernel """
    sqdist = np.sum(a**2,1).reshape(-1,1) + np.sum(b**2,1) - 2*np.dot(a, b.T)
    return (sigma ** 2) * np.exp(-.5 * sqdist / (l ** 2))

n = 50 # number of test points.
Xtest = np.linspace(-5, 5, n).reshape(-1,1) # Test points.
K_ = se_kernel(Xtest, Xtest, 1, 1) # Kernel at test points.
# draw samples from the prior at our test points.
L = np.linalg.cholesky(K_ + 1e-6*np.eye(n))
# m = np.dot(L, np.random.normal(size = (n, 10)))
# m.shape
f_prior = np.dot(L, np.random.normal(size=(n,10))) # + m
plt.plot(Xtest, f_prior)
plt.show()
```

## Rational Quadratic Kernel

```python
# decrease alpha: the lines become relative more constant, more jagged/wiggly i.e. less smooth. the range of the functions reduce and as alpha tends to
# zero, the curves become flat, horizontal i.e. constant lines.
# increase alpha: it approaches squared exponential.
# l is quite similar in its functioning to its squared exponential counterpart
# similar interactions
```

```python
def rq_kernel(a, b, l = 1, sigma = 1, alpha = 1):
    sqdist = np.sum(a ** 2, 1).reshape(-1, 1) + np.sum(b ** 2, 1) - 2 * np.dot(a, b.T)
    return (sigma ** 2) * (1 + sqdist/(2 * alpha * l ** 2)) ** -alpha
n = 50 # number of test points.
Xtest = np.linspace(-5, 5, n).reshape(-1,1) # Test points.
K_ = rq_kernel(Xtest, Xtest, 1, 1, 1) # Kernel at test points.
# draw samples from the prior at our test points.
L = np.linalg.cholesky(K_ + 1e-6*np.eye(n))
# m = np.dot(L, np.random.normal(size = (n, 10)))
# m.shape
f_prior = np.dot(L, np.random.normal(size=(n,10))) # + m
plt.plot(Xtest, f_prior)
plt.show()
```

```python
def periodic_tf(length_scale, period):
    """Periodic kernel TensorFlow operation."""
    amplitude_tf = tf.constant(1, dtype=tf.float64)
    length_scale_tf = tf.constant(length_scale, dtype=tf.float64)
    period_tf = tf.constant(period, dtype=tf.float64)
    kernel = tfk.ExpSinSquared(
        amplitude=amplitude_tf, 
        length_scale=length_scale_tf,
        period=period_tf)
    return kernel

def periodic(xa, xb, length_scale, period):
    """Evaluate periodic kernel."""
    kernel = periodic_tf(length_scale, period)
    kernel_matrix = kernel.matrix(xa, xb)
    matrix_numpy = kernel_matrix.numpy()
    print(kernel_matrix)
    return matrix_numpy
    # with tf.compat.v1.Session() as sess:
    #     return sess.run(kernel_matrix)
```

```python
def lin_kernel(a, b, sigma_b = 1, sigma_v = 1, c = 0):
    return sigma_b ** 2 * np.ones((a.size, b.size)) + sigma_v ** 2 * np.dot(a, b.T)
n = 50
Xtest = np.linspace(-5, 5, n).reshape(-1,1)
K_ = lin_kernel(Xtest, Xtest, 10, 10, 1)
L = np.linalg.cholesky(K_ + 1e-6*np.eye(n))
f_prior = np.dot(L, np.random.normal(size=(n,10)))
plt.plot(Xtest, f_prior)
plt.show()
```
